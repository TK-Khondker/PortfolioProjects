```python
# Impoting Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Configuring the plots


# Reading in data

df = pd.read_csv(r'C:\Users\TK\Downloads\movies.csv')
```

```python
# Looking at the data

df.head()
```

```python
# Looking for missing data

for col in df.columns:
    MissingDataPercent = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, MissingDataPercent))
```

```python
#Dropping rows with missing data

df = df.dropna()
```

```python
# Data types

df.dtypes
```

```python
# Converting column data types

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df.head()
```

```python
# Converting 'released' column in to a string and then extracting the year from it in to a new column to get correct year

#df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4] (Different way of doing it)

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df
```

```python
# Ordering the dataset in descending order by the 'gross'
df = df.sort_values(by = ['gross'], inplace = False, ascending = False)
#df.drop(columns = 'correctyear')
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Dropping any duplicates

df.drop_duplicates()
```

```python
# Checking for correlation between movie budget and the gross revenue using a scatter plot

plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget vs Gross Revenue')
plt.xlabel('Movie Budget')
plt.ylabel('GRoss Revenue')

plt.show()
```

```python
df.head()
```

```python
# Plotting budget vs gross using seaborn to show correlation (regression)

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color": "blue"}, line_kws = {"color": "green"})
```

```python
# Looking at correlation between the numeric fields

df.corr(numeric_only = True)
```

```python
# Trying different correlation methods (pearmon (default), kendall, spearman)

df.corr(method = 'pearson',numeric_only = True)
```

```python
df.corr(method = 'kendall', numeric_only = True)
```

```python
df.corr(method = 'spearman',numeric_only = True)
```

```python
# 

correlation_matrix = df.corr(method = 'pearson', numeric_only = True)
sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matric for Numeric Fields')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Making correlation easier to see

correlation_mat = df.corr(numeric_only = True)
corr_uns = correlation_mat.unstack()

sorted_corr = corr_uns.sort_values()

# Field pairs that have a high correlation '0.5' or higher
high_corr = sorted_corr[(sorted_corr) >= 0.5]
high_corr
```

```python
# Field pairs that have a low correlation 
low_corr = sorted_corr[(sorted_corr) < 0.5]
low_corr
```

```python
# Plotting Budget vs Score using seaborn to show correlation (regression)

sns.regplot(x = 'score', y = 'budget', data = df, scatter_kws = {"color": "blue"}, line_kws = {"color": "black"})
```
